# CPOA data structures for different operators

```python
import os
import json
from collections import defaultdict
import pandas as pd

operators = [item for item in os.listdir('data')]
print("Operators:")
_ = [print(o) for o in operators]
```

## metadata

```python
meta_d = {}
fields = []

for operator in operators:
    with open(f'data/{operator}/metadata.json') as f:
        data = json.load(f)
        fields += [key for key in data[0].keys()]
        meta_d[operator] = data
fields = list(set(fields))

fields_d = defaultdict(list)
for field in fields:
    present_ls = []
    for operator in operators:
        field_in_schema = field in meta_d[operator][0].keys()
        present_ls += [int(field_in_schema)]
    fields_d[field]=present_ls
            
meta_comparison_table = pd.DataFrame(fields_d,index=operators).transpose()
meta_comparison_table.sort_values(by=operators,ascending=False)
```
